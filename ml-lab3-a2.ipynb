```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df=pd.read_excel('embeddingsdata.xlsx')
# Assuming 'data' is your dataset and 'feature_of_interest' is the feature you want to analyze
feature_data = df['embed_0']  # Replace 'feature_of_interest' with your actual feature name

# Create a histogram
plt.hist(feature_data, bins=20)  # You can adjust the number of bins as needed
plt.xlabel('Feature Values')
plt.ylabel('Frequency')
plt.title('Histogram of Feature_of_Interest')
plt.show()

# Calculate the mean and variance
mean_value = np.mean(feature_data)
variance_value = np.var(feature_data)

print(f"Mean: {mean_value}")
print(f"Variance: {variance_value}")
```
