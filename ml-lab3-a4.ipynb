```python
import pandas as pd
from sklearn.model_selection import train_test_split
df=pd.read_excel("Dataset to B.tech.xlsx")
# Assuming 'df' is your DataFrame and 'label' is the name of the column to be dropped
X = df.drop(columns=['Label'])
y = df['Label']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(X_train)
print(X_test)
print(y_train)
print(y_test)
```
