```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_excel('embeddingsdata.xlsx')
# Assuming you have two feature vectors, feature_vector1 and feature_vector2
# Replace these with your actual feature vectors
feature_vector1 = df['embed_0'].to_numpy()
feature_vector2 = df['embed_1'].to_numpy()

# Create an array of r values from 1 to 10
r_values = np.arange(1, 11)

# Initialize an empty list to store the calculated distances
distances = []

# Calculate Minkowski distance for each value of r
for r in r_values:
    distance = np.linalg.norm(feature_vector1 - feature_vector2, ord=r)
    distances.append(distance)

# Plot the distances vs. r
plt.plot(r_values, distances, marker='o')
plt.xlabel('r Value')
plt.ylabel('Minkowski Distance')
plt.title('Minkowski Distance vs. r')
plt.grid(True)
plt.show()
```
